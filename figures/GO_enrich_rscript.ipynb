```r
library(ggplot2)
```

```r
df = read.table("go_enrich.Rin", sep = '\t', header = TRUE)
```

```r
head(df)
```

```r
subdf = df[df$Set == 'gastropods',]
p = ggplot(data=subdf, aes(x=reorder(Name, seqfreq), y=seqfreq*100, fill=-log2(adjpval))) + geom_bar(stat="identity") 
p = p + coord_flip()
p = p + theme_classic() 
p = p + labs(x=NULL, y='Frequency (%)', fill="-log2(P-value)") 
p = p + theme(text = element_text(size=6))
p = p + scale_fill_gradientn(limits=c(3, 22), colours = c('#fdd49e', '#fdbb84', '#fc8d59', '#ef6548', '#d7301f', '#b30000', '#7f0000'))
p = p + theme(legend.position="top")
p

ggsave('gastropods_goenrich_v3.pdf', device = "pdf", width = 3.5, height = 3.75, units = "in")
```
