```r
library(ggplot2)
```

```r
df = read.table("genome_coverage.Rin", sep = '\t', header = TRUE)
subdf = subset(df, rtype == 'illumina')
```

```r
max(subdf$count)
```

```r
gg = ggplot(data=df, aes(x=depth, y=count, color=rtype)) 
gg = gg + geom_line() 
gg = gg + theme_classic()
gg = gg + labs(x='Read depth', y='Count', color='Read type') 
gg = gg + scale_colour_brewer(palette = "Dark2")
gg = gg + geom_vline(xintercept = 20, color = "#d95f02")
gg = gg + geom_vline(xintercept = 86, color = '#1b9e77')

gg
```

```r
ggsave('genome_coverage.pdf', device = "pdf", width = 8, height = 2, units = "in")
```
