```r
library(ggplot2)
```

```r
df = read.table("assembly_stats.Rin", sep = '\t', header = TRUE)
```

```r
gg = ggplot(data=df, aes(x=length, y=genes/1000, size = n50, color=busco)) 
gg = gg + geom_point() 
gg = gg + theme_classic()
#gg = gg + scale_size(range = c(1, 10), breaks=c(100, 1000, 5000))
#gg = gg + scale_color_gradientn(colours = c('#7BBCE7', '#7EB2E4', '#88A5DD', '#9398D2', '#9B8AC4', '#9D7DB2', '#9A709E', '#906388', '#805770', '#684957', '#46353A'))
gg = gg + scale_color_gradientn(limits = c(85, 100), colours = c('#4D8AC6', '#4E96BC', '#549EB3', '#59A5A9', '#60AB9E', '#69B190', '#77B77D', '#8CBC68', '#A6BE54', '#BEBC48', '#D1B541', '#DDAA3C', '#E49C39', '#E78C35', '#E67932', '#E4632D', '#DF4828', '#DA2222'))     
#gg = gg + scale_color_gradientn(colours = c('#FEE391', '#FEC44F', '#FB9A29', '#EC7014', '#CC4C02', '#993404', '#662506'))
gg = gg + labs(x='Genome size (Mbp)', y='Protein-coding genes (x1000)', size='N50 (kbp)', color='Complete BUSCO (%)') 
gg = gg + theme(text = element_text(size=6))
gg = gg + geom_text(label=df$slug, size=1, color='#000000')
gg = gg + xlim(500, 1100)
gg = gg + ylim(10, 90)
gg = gg + theme(legend.key.size=unit(0.2, "cm")) 
gg = gg + theme(legend.key.width=unit(0.3, "cm")) 
gg
```
