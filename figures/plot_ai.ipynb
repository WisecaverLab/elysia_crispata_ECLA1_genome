```r
library(ggplot2)
library(ggrepel)
library(dplyr)
options(scipen = 999)
```

```r
df = read.table("ai_table.Rin", sep = '\t', header = TRUE)
```

```r
gg = ggplot(data=df, aes(x=normbitscoremetazoa, y=normbitscoreother, color=ai)) 
gg = gg + geom_point(alpha = 0.5) 
gg = gg + theme_classic() + coord_fixed()
gg = gg + geom_label_repel(data = subset(df, gene == "Ecla3748g428790"), aes(label = gene, size = NULL, color = ai), nudge_y = 0.5)
#gg = gg + geom_label_repel(data = subset(df, gene == "Ecla12387g116670"), aes(label = gene, size = NULL, color = ai), nudge_y = 0.5)
gg = gg + scale_color_gradientn(limits = c(-1, 1), colours = c('#4D8AC6', '#4E96BC', '#549EB3', '#59A5A9', '#60AB9E', '#69B190', '#77B77D', '#8CBC68', '#A6BE54', '#BEBC48', '#D1B541', '#DDAA3C', '#E49C39', '#E78C35', '#E67932', '#E4632D', '#DF4828', '#DA2222'))     
#gg = gg + geom_vline(xintercept = 2, color = "#000000")
#gg = gg + geom_hline(yintercept = -log10(0.01), color = '#000000')
#gg = gg + facet_grid(. ~ cat1)
#gg = gg + theme(legend.position = "none")
gg = gg + geom_abline(intercept = 0, slope = 1)
gg = gg + geom_density2d()
gg
```

```r
ggsave('alienindex.pdf', device = "pdf", width = 4, height = 5, units = "in")
```
