# Generate upset plot from OrthoFinder output

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from upsetplot import UpSet, from_memberships
```

```python
infile = 'input_for_upset_plot.txt'
```

```python
df = pd.read_csv(infile,sep="\t").set_index("Orthogroup")

group_dict={}

for index,row  in df.iterrows():
    for sp,count in row.items():
        if sp != "Total" and count != 0:
            group_dict.setdefault(index, []).append(sp)
```

```python
x=from_memberships(group_dict.values())
UpSet(x, subset_size='count',show_counts=True, sort_categories_by=None).plot();

figure = plt.gcf()
figure.savefig('upset_plot.pdf', format='pdf')
```
